```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load data
df = pd.read_csv("../Data/smart_admission_data.csv")
df.head()

# Basic structure
df.info()

# Descriptive statistics
df.describe()

# Check admission distribution
df['admitted'].value_counts(normalize=True).plot(kind='bar', color=['green', 'red'])
plt.title('Admission Outcome Distribution')
plt.xticks(ticks=[0, 1], labels=['Admitted', 'Not Admitted'], rotation=0)
plt.ylabel('Proportion')
plt.show()

#UTME score distribytion
sns.histplot(df['utme_score'], bins=30, kde=True, color='skyblue')
plt.title('UTME Score Distribution')
plt.xlabel('UTME Score')
plt.ylabel('Number of Applicants')
plt.show()

# Screening score distribution
sns.histplot(df['screening_score'], bins=30, kde=True, color='orange')
plt.title('Screening Score Distribution')
plt.xlabel('Screening Score')
plt.ylabel('Number of Applicants')
plt.show()

#Admission rate by faculty
faculty_admit = df.groupby('faculty')['admitted'].mean().sort_values(ascending=False)

sns.barplot(x=faculty_admit.values, y=faculty_admit.index, palette="viridis")
plt.title('Admission Rate by Faculty')
plt.xlabel('Admission Rate')
plt.ylabel('Faculty')
plt.xlim(0, 1)
plt.show()

#Relationship between Scores & Admission
# UTME Score vs Admission
sns.boxplot(x='admitted', y='utme_score', data=df, palette='Set2')
plt.title('UTME Score vs Admission Outcome')
plt.xticks([0, 1], ['Not Admitted', 'Admitted'])
plt.xlabel('Admission Outcome')
plt.ylabel('UTME Score')
plt.show()

# Screening Score vs Admission
sns.boxplot(x='admitted', y='screening_score', data=df, palette='Set1')
plt.title('Screening Score vs Admission Outcome')
plt.xticks([0, 1], ['Not Admitted', 'Admitted'])
plt.xlabel('Admission Outcome')
plt.ylabel('Screening Score')
plt.show()
```
